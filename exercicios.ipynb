## Explicação: Uso da técnica Self-Ask

Self-Ask é uma abordagem que divide uma pergunta complexa em subperguntas menores e resolve cada uma de forma iterativa, utilizando respostas anteriores para contextualizar a próxima etapa. No contexto do assistente virtual para a Câmara dos Deputados, essa técnica permite:

### Divisão do Problema:

Perguntas complexas, como "Qual deputado teve o maior gasto em Educação no último mês?", podem ser decompostas em:
- Quais deputados têm gastos em Educação?
- Qual foi o maior gasto registrado no último mês?
- Quem realizou esse gasto?

### Integração de Dados:

A técnica permite combinar diferentes fontes de dados vetorizados (deputados, despesas, proposições) para fornecer respostas precisas.

### Interatividade Inteligente:

O assistente pode "perguntar para si mesmo" questões intermediárias para refinar a resposta ao usuário.

## Comparação dos Modelos de Geração de Imagens

| **Modelo**           | **Arquitetura**                                                                                                                                     | **Vantagens**                                                                                                           | **Limitações**                                                                                                              |
| -------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------- |
| **Stable Diffusion** | Utiliza difusão latente, onde uma imagem ruidosa é refinada iterativamente. Baseia-se em modelos treinados para alta performance no espaço latente. | Altamente personalizável; menor consumo de memória comparado a modelos como DALL-E; integra facilmente estilos visuais. | Dependência de bons prompts; limitações em cenários muito abstratos ou com baixa criatividade intrínseca.                   |
| **DALL-E**           | Baseado em Transformer, combina texto com imagem para aprendizado e geração.                                                                        | Excelente para cenários de alta criatividade; grande diversidade de estilos e objetos.                                  | Consome mais memória; menor controle sobre ajustes específicos de estilo comparado ao Stable Diffusion.                     |
| **MidJourney**       | Utiliza técnicas proprietárias de geração; altamente otimizado para arte e criatividade visual.                                                     | Resultados artisticamente impressionantes; ideal para criar obras de arte estilizadas.                                  | Menor flexibilidade em prompts técnicos; dependência de plataforma própria; menos transparente para personalização técnica. |
